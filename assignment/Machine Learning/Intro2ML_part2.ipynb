### Q4: Explain the bias-variance tradeoff in machine learning. What is the relationship between bias and variance, and how do they affect model performance?


<b>bias-variance tradeoff: </b> For any model, we have to find the perfect balance between Bias and Variance. This just ensures that we capture the essential patterns in our model while ignoring the noise present it in. This is called Bias-Variance Tradeoff. It helps optimize the error in our model and keeps it as low as possible.


<br><br>
The relationship btw bias & variance is often referred as bias-variance tradeoff. Increasing the complexity of a model (e.g., by adding more features or increasing the degree of a polynomial regression) will decrease bias but increase variance, & vice versa. The goal of a good predictive model is to find a balance between bias and variance that minimizes the total error on the test data.

### Q5: Discuss some common methods for detecting overfitting and underfitting in machine learning models. How can you determine whether your model is overfitting or underfitting?


Detecting overfitting and underfitting:
1. Train/test split: Split data into Training & testing, then trained model on training set & evaluated on the test set if it perform well on Training but not in test then its likely overfitting and if it perform poorly on both then its underfitting

2. Cross-validation: This helps to prevent overfitting, by spliting data into k-folds and training our model with k-1 fold  and evaluating it on the remaining last fold.  

3. Learning curves: Ploting the Model performace on training and test sets. if Model overfits, then training error will be low & testing error will be high. if its underfits, then both error will be high 


4. Regularization:used to prevent overfitting by adding a penalty term to the loss function. This penalty term discourages the model from fitting the training data too closely and encourages it to generalize to new data.


### Q6: Compare and contrast bias and variance in machine learning. What are some examples of high bias and high variance models, and how do they differ in terms of their performance?


<b>Bias</b> refers to the difference between the predicted values of the model and the true values of the target variable
<br>eg., the model is making systematic errors that are consistently wrong. This can lead to poor performance on both the training data and the test data.
   
<br><b>Variance</b> refers to the variability of the model's predicted values for different training sets.
<br>eg., the model is making different errors for different subsets of the training data. This can lead to good performance on the training data, but poor performance on the test data.
<BR><BR>Relationship:
Increasing the complexity of a model (e.g., by adding more features or increasing the degree of a polynomial regression) will decrease bias but increase variance, and vice versa
