### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


```python
#Scatter plots are graphs that represent relationship btw two variables in data set
# in this both var. are put on X-Y plane and from that we analyze relationship 
#question part
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,marker="1")
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
<br>box2 = np.random.normal(90, 20, 200)

```python
# Using box plot we display data distribution using 5 summary - 
# 1 Minimum 
# 2 Q1 ( First Quartile)
# 3 median
# 4 Q3 (third Quartile)
# 5 Maximum
# through the box plot we got many info related to outliers and adding two or more 
# box plot we can compare them, there min, max, median, outliers values just by looking at 
# the visuallization without even looking at there values 


#Uses 
# These provides visual summary of data 
# Median gave us the average value of our data 
# Through this we can show skewness of the data
# dispersion or spread of data can be visualized
# this gave us idea about outliers (which are numerical distant data from rest of data)
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
fig = plt.figure(figsize= (8,6))

ax = fig.add_axes([0,0,1,1])
boxplot = ax.boxplot([box1, box2])
plt.show()
```
