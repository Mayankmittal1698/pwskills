Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris.head()
```

```python
# 1 distplot: help us to visualize the distribution of data. 
#              We can use these plots to understand the mean, median, range, variance, deviation, etc of the data
sns.histplot(iris.sepal_length, bins= 5) 
# here bin represent no. of fractions on plot
```

```python
# 2 Stripplot : A strip plot is drawn on its own. It is a good complement to a boxplot or violinplot in cases where all observations are
#  shown along with some representation of the underlying distribution. It is used to draw a scatter plot based on the category
sns.stripplot(x= 'sepal_length', y='species', data = iris) 
```

```python
# 3 Boxplot : usually used to get inside of the two field from dataset used to compare it with other field
sns.boxplot(y="sepal_length", x = "species", data = iris)
```

```python
# 4 Heatmap: graphical representation of data using colors to visualize the value of the matrix.
# more common values or higher activities brighter colors basically reddish colors are used and to represent less 
# common or activity values, darker colors are preferred.Also known as shading matrix.
import numpy as np
data = np.random.randint(low=1, 
                         high=100, 
                         size=(10,10))
print(data)
sns.heatmap(data)
```

```python
# 5 Scatterplot: used with several semantic groupings which can help to understand well in a graph. 
sns.scatterplot(x= 'sepal_length', y ='petal_length', data = iris , hue= 'species')
```


Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. <br>Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.


```python
# loading 'titanic' dataset 
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize = (10,5))

sns.boxplot(x= 'pclass', y ='age',data= titanic, ax = ax1)
sns.boxplot(x= 'pclass', y = 'fare', data = titanic, ax = ax2)
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights  = sns.load_dataset('flights')
flights.head()
```

```python
#pivot used to create a matrix where the rows represent the months, b
# columns represent the years,
# and the values represent the number of passengers.
flight_pivot = flights.pivot(index='month', columns='year',values='passengers')

sns.heatmap(flight_pivot)
```
